```python
import numpy as np
import matplotlib.pyplot as plt


def sinusoidal_position_encoding(seq_len, d):
    PE = np.zeros((seq_len, d))

    for pos in range(seq_len):
        for k in range(0, d, 2):
            freq = 1 / (10000 ** (2 * k / d))

            PE[pos, k] = np.sin(freq * pos)
            PE[pos, k + 1] = np.cos(freq * pos)

    return PE


# Example usage
PE_10x8 = sinusoidal_position_encoding(seq_len=10, d=8)
plt.figure(figsize=(12, 6))
plt.imshow(PE_10x8, aspect="auto", cmap="viridis")

plt.title("Sinusoidal Positional Encoding (Vaswani et al. 2017)")
plt.xlabel("Embedding Dimension")
plt.ylabel("Position")
plt.colorbar()
plt.show()
```
