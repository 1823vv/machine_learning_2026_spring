```python
import numpy as np
import matplotlib.pyplot as plt

positions = np.arange(0, 50)

sin_vals = np.sin(positions)
cos_vals = np.cos(positions)

plt.plot(positions, sin_vals, label="sin(i)")
plt.plot(positions, cos_vals, label="cos(i)")
plt.legend()
plt.title("Sine and Cosine as Position Functions")
plt.xlabel("Position i")
plt.ylabel("Value")
plt.show()
```
