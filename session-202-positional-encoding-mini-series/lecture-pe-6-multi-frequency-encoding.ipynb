```python
import numpy as np
import matplotlib.pyplot as plt

positions = np.arange(0, 100)

frequencies = [0.05, 0.1, 0.2, 0.5]

plt.figure(figsize=(10, 5))

for w in frequencies:
    signal = np.sin(w * positions)
    plt.plot(positions, signal, label=f"sin({w} * i)")

plt.title("Multi-Frequency Representation of Position")
plt.xlabel("Position i")
plt.ylabel("Value")
plt.legend()
plt.show()
```
